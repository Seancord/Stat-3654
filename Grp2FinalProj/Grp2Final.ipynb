```python
#5 All graphics were constructed and intended to be constructed in R. The powerpoint contains the graphics that were created using this
#code. The following sets of code are added here for completeness of the ipython notebook.
#Rule out trucks, by looking at passenger size
#Graph Y axis by passenger size, X being vehicle Class
#VClass: Small Pickup Trucks, Standard Pickup Trucks, 
#mydata[mydata&Vclass == mydata]
r("vehicleData$newData <- vehicleData$pv2 + vehicleData$hpv + vehicleData$pv4") 
r("vehicleData2 <- vehicleData[vehicleData$newData != 0,]")
r("plotFrame <- aggregate(newData ~ VClass, vehicleData2, FUN = 'median')")
r("plot(plotFrame$VClass, plotFrame$newData)")
r("par(cex.axis = .7, las = 2)")
r("title(main='Vehicle class vs Person space in hatchbacks', font.main=4)")
r("lines(plotFrame$vClass, plotFrame$newdata)")
```

```python
#the rest of the code pertains to the model for #6
r("vehicleData$gp <- runif(dim(vehicleData)[1])")
r("testSet <- subset(vehicleData, vehicleData$gp <= 0.1)")
r("trainSet <- subset(vehicleData, vehicleData$gp > 0.1)")
```
